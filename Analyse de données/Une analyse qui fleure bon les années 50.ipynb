## En examinant les valeurs propres et leur diagramme, déterminer les axes principaux qu'il faut garder dans les analyses.

**Attention** : nous voulons expliquer les marques par rapports aux attributs. Nos données sont donc les marques. Il faut transposer le tableau de données.

```python
from sklearn.decomposition import PCA 

fume_trans = fume.T
pca = PCA(n_components=5)
pca.fit(fume.T)
```

```python
print(pca.explained_variance_ratio_) #information véhiculée
print(pca.components_) #vecteur propre
```

Les 4 premiers axes expliquent 91% des données.

## Interprétation

### Les axes

Pour les axes retenus représeter sous forme d'un histogramme les corrélations aux anciens axes (`correlations[i].plot.bar()`)

```python
U = numpy.transpose(pca.components_) # vecteurs propres
C = pandas.DataFrame(fume_trans.dot(U)) # nouvelles coordonnées
I = pandas.DataFrame(pca.explained_variance_ratio_)  # information véhiculée

correlations = pandas.DataFrame([[C[facteur].corr(fume_trans[column]) for facteur in C] for column in fume_trans], index=fume_trans.columns)
correlations
```
